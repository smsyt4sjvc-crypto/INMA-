```python
import requests

# ---- paste YOUR paper keys here (local only — never share the secret) ----
API_KEY = "PASTE_YOUR_PAPER_KEY_ID"
SECRET  = "PASTE_YOUR_PAPER_SECRET_KEY"
# --------------------------------------------------------------------------

BASE = "https://paper-api.alpaca.markets/v2"
H = {"APCA-API-KEY-ID": API_KEY, "APCA-API-SECRET-KEY": SECRET}

if "PASTE_YOUR" in API_KEY or "PASTE_YOUR" in SECRET:
    print("!! Fill in API_KEY and SECRET above first (paper keys from the Alpaca dashboard).")
else:
    r = requests.get(BASE + "/account", headers=H, timeout=30)
    if r.status_code == 200:
        a = r.json()
        print("✅ CONNECTED to Alpaca PAPER")
        print("  account status :", a.get("status"))
        print("  cash           : $" + a.get("cash","?"))
        print("  buying power   : $" + a.get("buying_power","?"))
        print("  portfolio value: $" + a.get("portfolio_value","?"))
        print("  pattern day trader flag:", a.get("pattern_day_trader"))
        print("  (if cash is ~100000, the pipe works end to end.)")
    elif r.status_code in (401,403):
        print("❌", r.status_code, "unauthorized — the keys are wrong, or they're LIVE keys on the PAPER url.")
        print("   Regenerate PAPER keys in the dashboard (Paper toggle ON) and paste them again.")
    else:
        print("⚠️ unexpected:", r.status_code, r.text[:300])
```
